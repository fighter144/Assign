```python
#Q4  Using the given x and y data, generate a 3D surface plot to visualize the function?
```

```python
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
x,y=np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2+y**2))
data={"X":x.flatten(),"Y":y.flatten(),"Z":z.flatten()}
df=pd.DataFrame(data)
df
```

```python
fig=go.Figure(data=[go.Surface(x=x,y=y,z=z)])
fig.show()
```

```python
#Q3 Generate a Bokeh bar chart representing the counts of different fruits using the following datasets?
```

```python
fruits=["Apple","Orange","Banana","Pears"]
count=[20,25,30,35]
p = figure(x_range=fruits,height=350, title="Fruit Counts")
p.vbar(x=fruits, top=count, width=0.9)
p.xgrid.grid_line_color = None
p.y_range.start = 0
show(p)
```
